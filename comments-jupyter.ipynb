```python
import pandas as pd
```

```python
comments_iphone = pd.read_csv("comments_ml_iphone.csv")
comments_iphone = comments_iphone["comment"]
comments_iphone = pd.DataFrame(comments_iphone)
comments_iphone
```

```python
comments_samsung = pd.read_csv("comments_ml_samsung.csv")
comments_samsung = comments_samsung["comment"]
comments_samsung = pd.DataFrame(comments_samsung)
comments_samsung
```

```python
comments_xiaomi = pd.read_csv("comments_ml_xiaomi.csv")
comments_xiaomi = comments_xiaomi["comment"]
comments_xiaomi = pd.DataFrame(comments_xiaomi)
comments_xiaomi
```

```python
comments = pd.concat([comments_xiaomi, comments_samsung, comments_iphone], axis = 0)
comments = comments.dropna()
isnull = comments.isnull()
isnull.any().any()
```

```python
comments.to_csv("comments_concat.csv")
```
